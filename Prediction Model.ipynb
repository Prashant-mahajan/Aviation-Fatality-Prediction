```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.plotly as py
import plotly.graph_objs as go

# Machine Learning 
from sklearn.linear_model import LogisticRegression 
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron 
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics 
from sklearn.cross_validation import KFold
```

```python
path = './Dataset/df.xlsx'
```

```python
df = pd.read_excel(path)
df.head()
```

```python
df.info()
```

```python
# To check if there are any missing values in the dataframe and fill the missing data with the median value
df.isnull().any()
df = df.fillna(df.median())
```

```python
#df.isnull().any()
```

```python
df_training = df[:70000]
df_testing = df[70000:]
```

```python
df_testing = df_testing.drop(['Injury Severity','Country'],axis = 1)
X_train = df_training.drop(['Injury Severity','Country'],axis = 1)
Y_train = df_training['Injury Severity']
X_test = df_testing.copy()
```

```python
X_train.shape, Y_train.shape, X_test.shape
```

## Predictions

```python
## Logistic Regression 

logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```
